```python
dictt_traits=types_of_traits(traits)
```

```python
#Block to visualize the number of type of traits
import squarify 
labels = []
lenths = [len(dictt_traits[i]) for i in dictt_traits]
i = 0
for key in dictt_traits:
    label = key+'\n'+'('+str(lenths[i])+')'
    labels.append(label)
    i+=1
colors = [plt.cm.Spectral(i/float(len(dictt_traits))) for i in range(len(dictt_traits))]
plt.figure(figsize=(12,8),dpi=80)
squarify.plot(sizes = lenths,label = labels,color = colors,alpha = 0.8)
plt.title('A Map of  the number of type of traits')
plt.axis('off')
plt.show()
```
